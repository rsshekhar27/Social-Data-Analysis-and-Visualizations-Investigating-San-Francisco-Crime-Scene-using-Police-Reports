```python
# load the full file into df
df = pd.read_csv('Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv')

df.head()
```

```python
#define the latitude and longitude of San Francisco
lan = 37.77919
lon = -122.41914

# plot San Francisco's map
m = folium.Map([lan, lon], zoom_start=13)
m
```

```python
# make sure the date formate is right, NOTE: it takes a long time to run
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
#select the time peroid from June, 2016 to July, 2016
start_date = '2016-06-01'
end_date = '2016-7-31'

# load the certain peroid of data to df_part
df_part = df[(df['Date'] > start_date) & (df['Date'] <= end_date)]

# Sort the arrest by two categories: DRUG/NARCOTIC and LARCENY/THEFT within the time peroid
df_DN=df_part[(df_part['Category'] == 'DRUG/NARCOTIC') & (df_part['Resolution'] == 'ARREST, BOOKED')]
df_LT=df_part[(df_part['Category'] == 'LARCENY/THEFT') & (df_part['Resolution'] == 'ARREST, BOOKED')]
```

```python
# check the filtered (DRUG/NARCOTIC) data
df_DN.head()
```

```python
# chenck the filtered LARCENY/THEFT data
df_LT.head()
```

## Draw a little circle for each arrest in the category 'DRUG/NARCOTIC'

```python
# mark each category which is 'DRUG/NARCOTIC' as a point and plot it on the map
m1 = folium.Map([lan, lon], zoom_start=13)
for index, row in df_DN.iterrows():
    folium.CircleMarker([row['Y'], row['X']],# note in the datafram Y is lat and X is lon
                        radius=15,
                        popup=row['Resolution'],
                        fill_color="#3db7e4", # divvy color
                       ).add_to(m1)
m1
```

## Draw a little circle for each arrest in the category 'LARCENY/THEFT'

```python
# mark each category which is 'LARCENY/THEFT' as a point and plot it
m2 = folium.Map([lan, lon], zoom_start=13)
for index, row in df_LT.iterrows():
    folium.CircleMarker([row['Y'], row['X']],# note in the datafram Y is lat and X is lon
                        radius=15,
                        popup=row['Resolution'],
                        fill_color="#3db7e4", # color
                       ).add_to(m2)

m2
```

## Plot the heat map for each arrest in the category 'DRUG/NARCOTIC'

```python
# convert to (n, 2) nd-array format for heatmap
df_DN_Arr = df_DN[['Y', 'X']].values
m1 = folium.Map([lan, lon], zoom_start=13)


# plot heatmap

heat=plugins.HeatMap(list(df_DN_Arr), radius=20)
m1.add_child(heat)
m1
```

## Plot the heatmap for each arrest in the category 'LARCENY/THEFT'

```python
# convert to (n, 2) nd-array format for heatmap
df_LT_Arr = df_LT[['Y', 'X']].values

# plot heatmap
m2 = folium.Map([lan, lon], zoom_start=13)
heat=plugins.HeatMap(list(df_LT_Arr), radius=20)
m2.add_child(heat)
m2
```
